```python
import os
import numpy as np
import keras
import tensorflow
from tensorflow.keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img
```

```python
from google.colab import drive
drive.mount("/content/drive", force_remount=True)
```

```python
def data_augmentation(img_path, save_path):
    # Load the image
    img = np.array(tensorflow.keras.preprocessing.image.load_img(img_path))

    # Define the data generator
    datagen = ImageDataGenerator(
        rotation_range=45,
        width_shift_range=0.2,
        height_shift_range=0.2,
        horizontal_flip=True,
        vertical_flip=True,
        fill_mode='constant',
        rescale=1./255)

    # Generate the augmented images
    i = 0
    num_images = 8  # total images you want per original image
    batch_size = 5   # generate 5 images per batch

    for batch in datagen.flow(
        np.expand_dims(img, axis=0),
        batch_size=batch_size,
        save_to_dir=save_path,
        save_prefix='aug',
        save_format='jpg'
    ):
        i += batch.shape[0]  # batch.shape[0] is the number of images in this batch
        if i >= num_images:
            break
```

```python
# Input path
input_path = "/content/drive/MyDrive/Augmentation/tomato mold"

# Output path
output_path = "/content/drive/MyDrive/Augmentation/tomato mold"

# Loop through all images in the input path
for filename in os.listdir(input_path):
    img_path = os.path.join(input_path, filename)
    data_augmentation(img_path, output_path)
```
